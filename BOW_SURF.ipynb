```python
import cv2
import os
import numpy as np
from scipy.spatial import distance

imgs2Keypoints = {}
kmeansTrainer = cv2.BOWKMeansTrainer(10)

imagepath = '/home/topica/Downloads/img/transfer'
list_image =  os.listdir(imagepath)
for i in range(len(list_image)):
    list_image[i] = os.path.join(imagepath, list_image[i])

for pathToImage in list_image:
    sift = cv2.xfeatures2d.SURF_create()
    img = cv2.imread(pathToImage)
    kp, des = sift.detectAndCompute(img, None)
    des = np.float32(des)
    kmeansTrainer.add(des)
    imgs2Keypoints[pathToImage] = kp 
vocabulary = kmeansTrainer.cluster()
surf2 = cv2.xfeatures2d.SURF_create()
bow_ext = cv2.BOWImgDescriptorExtractor(surf2, cv2.BFMatcher(cv2.NORM_L2))
bow_ext.setVocabulary(vocabulary)

#returns descriptor of image at pth
def feature_extract(pth):
    img = cv2.imread(pth)
    return bow_ext.compute(img, imgs2Keypoints[pth])[0]

desc1 = feature_extract('/home/topica/Downloads/img/transfer/4_gray.png')
result = []
for pathToImage in list_image:
    desc2 = feature_extract(pathToImage)
    dst = distance.euclidean(desc1, desc2)
    result.append(dst)
index = sorted(range(len(result)), key=lambda k: result[k])
for i in range(len(index)):
    print(list_image[index[i]])
    print(result[index[i]])
```
